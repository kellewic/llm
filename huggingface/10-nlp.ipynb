## Natural Language Processing

### Masking

```python
from transformers import pipeline
```

```python
fill_mask = pipeline("fill-mask", model="bert-base-uncased")
```

```python
fill_mask("I am going to [MASK] later today.")
```

```python
fill_mask("The sun is [MASK].")
```

### Named Entity Recognition (NER)

```python
ner = pipeline("ner", model="dslim/bert-base-NER")
```

```python
ner("Paris is in France which is in Europe.", grouped_entities=True)
```

```python
## Doesn't recognize "mazda"
ner("My name is Jose and I drive a green mazda and work at Google.", grouped_entities=True)
```

```python
## Let's change the model so it recognizes "mazda"
del ner
ner = pipeline("ner", model="dslim/bert-base-NER-uncased")
ner("My name is Jose and I drive a green mazda and work at Google.", grouped_entities=True)
```

### Question & Answering

```python
qa = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
qa(question="What is hungry?", context="The cat that is sitting on the mat is hungry.")
```

```python
qa(question="What tasks can an LLM do?", context="LLMs are trained on massive amounts of data. They are designed to perform tasks like text generation, translation, question answering, and more, by learning patterns.")
```
